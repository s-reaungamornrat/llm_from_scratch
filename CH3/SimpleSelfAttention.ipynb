```python
import torch

inputs=torch.tensor( # 6 tokens long sentence, each token is 3D embedding
    [[0.43,0.15,0.89],
     [0.55,0.87,0.66],
     [0.57,0.85,0.64],
     [0.22,0.58,0.33],
     [0.77,0.25,0.10],
     [0.05,0.80,0.55]]
)
query=inputs[1]
print(f"{query.shape=}")
attn_scores2=torch.matmul(inputs, query[:,None]).squeeze(-1)
print(f"{attn_scores2.shape=}\n{attn_scores2}")
```

```python
attention=torch.matmul(inputs, inputs.T) # inputs @ inputs.T
print(f"{attention.shape}\n{attention}")
normalized_attention=attention/attention.sum(dim=1,keepdim=True)
print(f"{normalized_attention.shape}\n{normalized_attention}")
```

```python
# each row i-th represents the attention of token i-th to other tokens in this sentence
soft_max_attention=torch.softmax(attention, dim=1) 
print(f"{soft_max_attention.shape=}\n{soft_max_attention}")
soft_max_attention.sum(dim=1)
```

```python
query=inputs[1]
print(f"{query.shape=}\n{query}")
context_vec2=torch.zeros(query.shape)
for i, x_i in enumerate(inputs):
    print(i, f" {x_i.shape=}, {x_i}")
    context_vec2+=soft_max_attention[1,i]*x_i

print(f"\n{context_vec2=}")
```

```python
# attention of each row i-th applied to all tokens in each embedding dimension
context_vectors=torch.matmul(soft_max_attention, inputs) # soft_max_attention @ inputs
print(f"{context_vectors.shape=}\n{context_vectors}")
```
