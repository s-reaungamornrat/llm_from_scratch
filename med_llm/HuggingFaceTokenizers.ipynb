```python
from transformers import AutoModelForCausalLM, AutoTokenizer
```

```python
text=("Transthyretin Amyloid Cardiomyopathy (ATTR-CM): A fatal, underdiagnosed condition caused by abnormal protein buildup "
      "that stiffens the heart muscle, typically affecting older adults")
```

```python
model_path="OpenMedZoo/MedGo"
tokenizer=AutoTokenizer.from_pretrained(model_path, trust_remote_code=True)
inputs=tokenizer(text, return_tensors="pt")
token_ids=inputs["input_ids"]
print(f"{token_ids=}")
decoded_text = tokenizer.decode(token_ids)
print(f"{decoded_text=}\n")

vocab_dict = tokenizer.get_vocab()
print(f"\n{tokenizer.vocab_size=}, {len(tokenizer)=}\n{list(vocab_dict.items())[30000:30010]=}\n")
for token_id in token_ids.squeeze():
    print(f"{token_id:<10}|\t{tokenizer.decode([token_id]):<10}")
```
