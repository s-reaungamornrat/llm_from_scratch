```python
text=open('../taylorswift.txt', 'r', encoding='utf-8').read()
print(len(text))
```

```python
main_dir=r'/Users/reaungamornrat.sureerat/data/BPE/minbpe'
```

```python
basic_tokenizer=BasicTokenizer()
basic_tokenizer.train(text, vocab_size=512, verbose=False)
input_text=("Transthyretin Amyloid Cardiomyopathy (ATTR-CM): A fatal, underdiagnosed condition caused by abnormal protein buildup "
      "that stiffens the heart muscle, typically affecting older adults")
ids=basic_tokenizer.encode(input_text)
print(f"{ids=}")
print("\n", basic_tokenizer.decode(ids))
```

```python
re_tokenizer=RegexTokenizer()
re_tokenizer.register_special_tokens(special_tokens={"<|endoftext|>":512, "<|beginoftext|>":513})
re_tokenizer.train(text, vocab_size=512, verbose=False)

input_text=("<|beginoftext|>Transthyretin Amyloid Cardiomyopathy (ATTR-CM): A fatal, underdiagnosed condition caused by abnormal protein buildup "
      "that stiffens the heart muscle, typically affecting older adults<|endoftext|>")
ids=re_tokenizer.encode(input_text, allowed_special={"<|endoftext|>", "<|beginoftext|>"} )
print(f"{ids=}")
print("\n", re_tokenizer.decode(ids))
```
