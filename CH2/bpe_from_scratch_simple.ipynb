The goal of the byte-pair encoding (BPE) algorithm is to build a vocabulary of commonly occurring subwords

```python
import tiktoken

%load_ext autoreload
%autoreload 2

from bpe_simple import BPETokenizerSimple
```

```python
text="This is some text"
byte_ary=bytearray(text, 'utf-8') # one byte per character
print(len(byte_ary), len(text))
print(byte_ary)
```

```python
# when we call list() on a byte array, each byte/character is treated as an individual element, and the result is a list of integers corresponding
# to the byte values
ids=list(byte_ary)
print(len(ids), ids)
```

```python
gpt2_tokenizer=tiktoken.get_encoding("gpt2")
gpt2_tokenizer.encode(text)
```

```python
for i in range(300):
    decoded=gpt2_tokenizer.decode([i])
    print(f"{i}: '{decoded}'")
# the entries 256 and 257 are not single-character values but double-character values (whitespace+letter) which is a shortcoming of GPT2 BPE
# tokenizer (which has been improved in GPT4 tokenizer)
```

```python
import os
import requests

main_dir='/Users/reaungamornrat.sureerat/data/BPE/pracetice'
file_path=f"{main_dir}/the-verdict.txt"

if not os.path.exists(file_path):
    url = ("https://raw.githubusercontent.com/rasbt/"
           "LLMs-from-scratch/main/ch02/01_main-chapter-code/"
           "the-verdict.txt")
    response = requests.get(url, verify=False)
    with open(file_path, 'wb') as f: f.write(response.content)

with open(file_path, "r", encoding="utf-8") as f:  text = f.read()
```

```python
replace_space="Ġ"

tokenizer=BPETokenizerSimple(replace_space=replace_space)
assert tokenizer.find_freq_pair([]) is None
assert tokenizer.find_freq_pair([42]) is None

tokenizer.train("", vocab_size=270)
tokenizer.train("H", vocab_size=270)
tokenizer.train("He", vocab_size=270)
print("Edge-case check passed")

tokenizer.train(text, vocab_size=1000, allowed_special={"<|endoftext|>"})
print(f"{len(tokenizer.vocab)=}, {len(tokenizer.bpe_merges)=}")
print(list(tokenizer.vocab.items())[-10:])
```
